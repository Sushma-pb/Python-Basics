# Loc()

```python
#access group of rows and column BY LABELS   like thier names  (location)
```

```python
#Basically loc method returns desired location values in data frame provided the input of that location is given
```

```python
import pandas as pd
```

```python
df= pd.read_csv("Fortune_11.csv")
df
```

```python
df.loc[0]     #label meaning index   Fetching 0th row with column name
```

```python
df.loc[1] 
```

```python
df.loc[[0,5]]       #giving input as list of label to print multiple rows
```

```python
#this has returned  0 and 5 index values 
```

```python
df.loc[[2,4]]   
```

```python
#if we want the 4 th value of Name
```

```python
df.loc[4,"Name"] #if we want the 4 th value of Name Column
```

```python
df
```

```python
#input as slice object with labels
```

```python
df.loc[0:1,"Name"]    #we have fetched 0 to 3 values of Name column   Slicing is inclusive
```

```python
df.loc[0:3,"Growth"]
```

```python
df
```

```python
#in Id column we want to fetch values less than 5
```

```python
df.loc[df["ID"]<5]    #we can apply condtions also to fetch desired values
```

```python
#now lets say we dont want the entire data set rather we need only growth of the less than 5 
```

```python
df.loc[df["ID"]<5,["Growth"]] 
```

# iloc()

```python
#Integer location Based method Meaning you ccan access rows and column by indexing or givnig thier a number 
```

```python
#generally we perform data selection operation by loc and iloc
```

```python
df
```

```python
df.iloc[0]   #returned the 0 index of the data frame
```

```python
df.iloc[[0]]   #to print it out systematically
```

```python
df.iloc[:,0]    #all rows and oly the 0 column
```

```python
df.iloc[:,3]
```

```python
df.iloc[[0,1]] #returned 0 and 1 index
```
