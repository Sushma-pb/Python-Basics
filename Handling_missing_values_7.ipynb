```python
#dropna()  let us understand the parameters of dropna() 
```

```python
   #it has axex as parameters axis = 0 drops the row containig nan values
               #axis =1 drops the column containig the missing value
    
                #bydefault value will be axis=0 meaning it will drop the row containig missing value
```

```python
df
```

```python
#here we observe that EXPENSES column has nul value 
```

```python
df.dropna(axis = 1)     #this will drop that particular column which has null value
                        #in this case EXpense column has been dropped
```

```python
# after axis the next parameter will study is HOW in How parameter we give two types of input "any" or "all"
#and the default value is "any"
```

```python
df.dropna(how = "any")   #it has droped only that row which had nan value
                         #this is quite similar to dropping only rows by giving axis =0 ,
                          #dont worry about that will see the future use of it   
```

```python
df.dropna(axis=1,how = "any") # now what will happen is it will drop that particular column which has nan values 
```

```python
df.dropna(how = "all")  #This will drop the rows and column only if there are all the missing values 
                        #meaning all values are missing then only this will work
```

```python
df   #looks same as above
```

```python
#if you have a column of all null values or rows of all null values then this will drop that particular row or column
```

```python
#now let us understan the thesh or threshold parameter of the dropna()
```

```python
df.dropna(thresh = 1) #meaning it will print only those rows or column which has atleas one notnull value
                       #meaning if we have on correct value it will print 
                        #lets say we have a row or coulmn of nan values it will not print 
```

```python
df.dropna(thresh = 8)   #look here it has dropped that row which didnot have 8 values
```

```python
#let us study next parameter which is subset
```

```python
df.dropna(subset= ["Expenses"])    #it dropped that row which had nan value
```

```python
#let us study the inplace parameter 
```

```python
df.dropna(inplace = True )   #this will not create a new data frame rather update the same data frame
df
```
