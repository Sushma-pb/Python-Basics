# fillna()


```python
#fill na fills the NAN value with given values input
```

```python
import pandas as pd
```

```python
df = pd.read_csv("Fortune_11.csv")
```

```python
df
```

```python
#let us look at the Parameter of fillna method
```

```python
df.fillna(0)   #value parameter if you give the value = "0" the nan values will be replaced by o
                #this value is filled in entire data set
```

```python
#look at the 8 row and in Expense column
```

```python
df.fillna(5) 
```

```python
#now let say you have to fill  a particular data  in particular column
```

```python
df.fillna({"Expenses":"none"})   #({"Expenses":"none"},{"Profit = 30"},{}) You can extend this library for different column
```

```python
#here look at the expense column i have filled a string values 
#by doing this you can fill integer or string by calling the particular column in the dictionary
```

```python
#let us see the next parameter of fillna()
#method 
```

```python
df.fillna(method = "ffill") #meaning forward fill the previous value will be filled in the nan value Previous value
```

```python
df.fillna(method = "bfill")
```

```python
df.fillna(axis = 0,method = "bfill")   #with axis we have to use any one like value or method parameter other wise
                                           # it will through an error
```

```python
df.fillna(axis = 1,method = "bfill")
```

```python
df.fillna(limit = 1,value = 0)       #use to fill specific NAN values  This will fill 0 in only one position
                                     #meaning how much you limit it will fill only those values lets say limit = 5 it will
                                    #fill only 5 values,,...here also method and value parametr is compulsory
```

```python
df.fillna(limit = 2,value = 1) 
```

```python
df.fillna(limit = 2,method = "ffill")  #lets use method parameter
```

```python
 df.fillna(value = 5,inplace = True)  #inplace = True will modify that data frame it will not create a new data frame'
df
```
