# Data visualization Scatter plot  class 1

```python
import seaborn as sns  # plot data
```

```python
import pandas as pd # import data sheet
```

```python
sales = pd.read_excel('sales.xls') # read excel sales file
```

```python
sales.head()
```

```python
sns.relplot('Sales','Profit', data=sales) # showing sales and profit data in scatter plot usin sns 
```

```python
sns.relplot('Sales','Profit', data=sales, hue ='Order Priority') # using hue better under standing of sales and profit data
```

```python
sns.relplot('Sales','Profit', data=sales, hue ='Order Priority', style = 'Order Priority') 
# using style for under stanging of data
```

```python
sns.relplot('Sales','Profit', data=sales, hue ='Order Priority', style = 'Ship Mode') 
# using ship moe in style current situation of data
```

```python
sns.relplot('Sales','Profit', data = sales, size ='Discount') # using discount in size incresing size of point
```

```python
sns.relplot('Sales','Profit', data = sales, size ='Discount', sizes =(25,200)) # rang the size of point much better to see
```

# Data visualization Line Plot class 2

```python
sales.head()
```

```python
sns.relplot('Order Date','Sales',data=sales, kind='line') # line plot
```

```python
sales['year']=sales['Order Date'].dt.year
```

```python
sales.year.unique()
```

```python
sns.relplot('year','Sales',data=sales,kind='line')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci="sd")
```

```python
sns.relplot('year','Sales',data=sales,kind='line', estimator=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line', estimator =sum)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style='Product Category')
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style='Product Category',ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col='Product Category',ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col='Region',ci=None)
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',col='Region',ci=None,col_wrap=2)
```

# Class 4

```python
sales.head()
```

```python
sns.catplot('Ship Mode','Sales',data=sales) #  catagorical data  using scatter plot
```

```python
sns.catplot('Ship Mode','Sales',data=sales,jitter =False) # using jitter to dont show the magnitude
```

```python
sns.catplot('Ship Mode','Sales',data=sales.query('Profit > 1000'),jitter =False)
```

```python
sns.catplot('Ship Mode','Sales',data=sales.query('Profit > 1000'),kind='swarm')
```

```python
sns.catplot('Sales','Ship Mode',data=sales.query('Profit > 1000'),kind='swarm') # making vertical plot
```

#  class 5 Box Plot Chart

```python
sales.head()
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,kind='box') # using box plot chart
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',hue='Ship Mode',data=sales,kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',hue='Ship Mode',hue_order=['Delivery Truck','Regular Air','Express Air'],
            data=sales,kind='box')
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',row='Ship Mode',hue_order=['Delivery Truck','Regular Air','Express Air'],
            data=sales,kind='box') # make single chart deffrent ship made in row
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',col='Ship Mode',hue_order=['Delivery Truck','Regular Air','Express Air'],
            data=sales,kind='box') # make single chart deffrent ship made in colum
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',row='Ship Mode',col='Customer Segment',hue_order=['Delivery Truck','Regular Air','Express Air'],
            data=sales,kind='box') # make single chart deffrent ship made and coustomer segment in row and colum
```
