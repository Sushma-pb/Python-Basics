#  class 6 violin plot

```python
import seaborn as sns
```

```python
import pandas as pd  # to import data sheet
```

```python
sales = pd.read_excel('sales.xls') # read excel sales file
```

```python
sales.head(2)
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',data=sales,kind='violin') #using violin plot better to visualizing data
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',hue='Ship Mode',data=sales,kind='violin') #using diffrent ship mode in chart
```

```python
sns.catplot(x='Product Category',y='Shipping Cost',hue='Ship Mode',inner='stick',data=sales,kind='violin') 
# using inner stick beter to visualizing a data
```
