# Seaborn 

```python
import seaborn as sns    
import pandas as pd     #To import dataset
```

```python
sales=pd.read_excel('sales.xls') # Importing sales dataset
sales.head()    
```

```python
sns.relplot('Sales','Profit',data=sales)    #it shows Relationship btw  2 column parameters [sales(X) & profit(Y)] 
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Order Priority')  #hue=color code to undrastand & interaction of dataset
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Order Priority',style='Order Priority')   #style=it shows the shapes 
```

```python
sns.relplot('Sales','Profit',data=sales,hue='Order Priority',style='Ship Mode')   #ship mode=it indicates respective shipping types[regular,express air,delivery truck]
```

```python
sns.relplot('Sales','Profit',data=sales,size='Discount')  
```

```python
sns.relplot('Sales','Profit',data=sales,size='Discount',sizes=(25,200)) #sizes =to increse in discount sizes  
```
