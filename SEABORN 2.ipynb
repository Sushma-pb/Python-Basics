# seaborn line plot

```python
import seaborn as sns
import pandas as pd
```

```python
sales=pd.read_excel('sales.xls') # Importing sales dataset
sales.head()    
```

```python
sns.relplot('Order Date','Sales',data=sales,kind='line') 
#here order date is X axis,Sales is Y axis & we use kind= line to show the line chart
```

```python
sales['year']=sales['Order Date'].dt.year #here .dt.year is used to extract year values
```

```python
sales.year.unique() #unique is the method to print the unique values of the year in order
```

```python
sns.relplot('year','Sales',data=sales,kind='line')   #here observe,what r ol changes in  the graphs (above &  below) carefully.
#in below graph we can observe d single line ,dt is our calculated mean & shaded area is the confidential interval (+95% & -95%) 
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci=None) #here ci=confidential interval & the value "None" value removes the ci
```

```python
sns.relplot('year','Sales',data=sales,kind='line',ci="sd")  # sd =standard deviation 
```

```python
sns.relplot('year','Sales',data=sales,kind='line',estimator=None)  #estimator gives the exact values dt,what happened in the perticular year  
```

```python
sns.relplot('year','Sales',data=sales,kind='line',estimator =sum) #estimator = sum ,here "sum" adds  all the values & it gave a result along with ci.
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode') #Ship Mode //devides our values with different ship modes based on d value(hue)which 
```

```python
sns.relplot('year','Sales',data=sales,kind='line',hue='Ship Mode',style='Product Category ')  #
```
