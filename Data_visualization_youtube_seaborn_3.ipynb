#  Class 9 Distribution Plot

```python
import pandas as pd # to import data sheet
```

```python
import seaborn as sns # import plot
```

```python
import numpy as np # import numpy  package
```

```python
sales=pd.read_excel('sales.xls') #read sales file
```

```python
sales.head(2)
```

```python
sns.distplot(sales.Sales) #  sales to use in  distribution plot using distplot
```

```python
sns.distplot(sales.Profit) # profit to use in distribution plot
```

```python
sns.distplot(sales.Profit, hist=False) # only showing histogram
```

```python
sns.distplot(sales.Profit, hist=False, rug=True) # using rug to check how mony data point we having in x axes
```

```python
sns.distplot(sales.Profit,kde=False,rug=True,bins=100)
```

```python
sns.distplot(sales.Profit[(sales.Region == "West" )& (sales.Profit>0)], hist=False, rug=True)
# check it in particuler region of profit gretar then 0
```

```python
sns.distplot(sales.Profit)
```

```python
sns.kdeplot(sales.Profit,bw=1000)# check bandwidth in kdeplot 
```

```python
from scipy import stats  # import scipy library for gamma,beta,lot of distribution
```

```python
a= np.random.gamma(3,size=200)
```

```python
sns.distplot(a)
```

```python
sns.distplot(a,fit=stats.gamma) # fit parameter help to fit distrubution
```

```python
b= np.random.beta(7,7,size=200) # using beta parameter to distribution data
```

```python
sns.distplot(b,kde=False,fit=stats.beta)
```
